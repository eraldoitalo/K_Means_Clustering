```python
# KMeans_clutering
# importando libs
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Gerar dados simulados com 3 grupos
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
# visualizar os dados antes do agrupamento
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Dados Originais (sem agrupamento)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Aplicar K-Means com 3 grupos
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

```python
# Obter os rotulos de grupo e os controides
y_kmeans = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

```python
# Visualizar os argumentos encontrados
plt.figure(figsize=(6, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_kmeans, palette="deep", s=50)
plt.scatter(centroids[:, 0 ], centroids[:, 1], c='black', s=200, alpha=0.75, marker='X', label='Centroides')
plt.title("Argumentos com K-Means (K=3)")
plt.legend()
plt.show()
```
