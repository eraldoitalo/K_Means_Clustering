```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Configurar a visualização
sns.set(style="darkgrid")
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
# Geração de dados sintéticos (300 pontos, 2 dimensões, 4 clusters)
X, y_true = make_blobs(n_samples=350, centers=5, cluster_std=0.6, random_state=25)
```

```python
# Visualizção inicial dos dados (sem clustering)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Treinamento do modelo K_Means com K=4
kmeans = KMeans(n_clusters=5, init='k-means++', n_init=10, max_iter=300, random_state=25)
kmeans.fit(X)
```

```python
# Resultado dos modelos
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
inertia = kmeans.inertia_
```

```python
print(f"Inércia final: {inertia:.2f}")
```

```python
# Visualização dos clusters encontrados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c="black", s=200, alpha=0.75, marker='X', label="Centroids")
plt.title("Clusters indentificados pelo K-Means")
plt.legend()
plt.show()
```
